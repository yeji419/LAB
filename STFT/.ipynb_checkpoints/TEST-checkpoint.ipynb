```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
import math
import import_ipynb
import soundfile as sf
from STFTAlgorithm import STFT, ISTFT, IMSTFT, GriffinLim
```

```python
sig, sr = librosa.load('my voice.wav' , sr = 16000)
```

```python
window_length = 512
shift = 256
```

```python
stft = STFT(sig, window_length, shift)
```

```python
print(stft.shape)
```

```python
half_stft = stft[:,:257].T
log_spectrogram = librosa.amplitude_to_db(np.abs(half_stft))

librosa.display.specshow(log_spectrogram, sr=sr, hop_length=shift)
```

```python
lib_stft = librosa.stft(sig, n_fft=window_length, hop_length=shift)
log_spectrogram1 = librosa.amplitude_to_db(np.abs(lib_stft))
librosa.display.specshow(log_spectrogram1, sr=sr, hop_length=shift)
```

```python
Mask = np.exp(np.pi*1j)
mstft = Mask * stft
```

```python
result = GriffinLim(mstft, window_length, shift)
```

```python
istft = librosa.istft(stft)
```

```python
plt.plot(sig, label='Original Signal')
plt.plot(result,label='Griffin-Lim Algorithm')
plt.plot(imstft, label='IMSTFT')
plt.plot(istft, label ='librosa ISTFT')
plt.legend()
```
